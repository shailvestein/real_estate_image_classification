```python
def load_image(img_path, size=None):
    """this function will load image from disk
    size: if size is given it will return resized image of given size else return original size image
            default is 'None'"""
    ### reading image
    image = cv.imread(img_path, cv.IMREAD_UNCHANGED)
    ### changing the channels from BGR to RGB
    image = image[:,:,::-1]
    ### if size in None return original image
    if size == None:
        return image
    ### if size is given return resized image
    else:
        resized_image = cv.resize(image, size, interpolation=cv.INTER_AREA)
        return resized_image
```

```python
def get_dimensions(image_paths):
    """This function will return the dimensions of the input iname path"""
    ### empty set variables to store height, width and channel of images
    heights, widths, channels = list(), list(), list()
    
    for img_path in tqdm(image_paths):
        ### loading image
        img = load_image(img_path)
        ### shape of image array
        shape = img.shape
        ### checing if image haing 3 channel
        if len(shape) > 2:
            height = shape[0]
            width = shape[1]
            channel = shape[2]
            ### appending dimension to the set
            heights.append(height)
            widths.append(width)
            channels.append(channel)
        ### if image having 2 channel
        else:
            height = shape[0]
            width = shape[1]
            ### appending dimensions to the set
            heights.append(height)
            widths.append(width)
            
    return heights, widths, channels
```

<p style="font-size:25px; color:balck; font-weight:bold; font-style:italic;">7.&ensp;Preprocessing</p>

<p style="font-size:17px; color:Blue; font-style:italic;">Contrast limited adaptive histogram equalization (CLAHE) is used for improve the visibility level of foggy image or video. Please find more details about CLAHE <a href="https://en.wikipedia.org/wiki/Adaptive_histogram_equalization#:~:text=Contrast%20Limited%20AHE%20(CLAHE)%20is,slope%20of%20the%20transformation%20function.">here.</a></p>

```python
def rescale(image):
    """this function will rescale the image and will be used if requires"""
    image = np.array(image)/255
    return image
```

```python
def get_enhanced_image(colorimage, clipLimit=2, tileGridSize=(8,8)):
    """this function will return enhanced image using technique called CLAHE"""
    ### sorurce: https://www.freedomvc.com/index.php/2021/09/19/color-image-histogram-clahe/
    # We first create a CLAHE model based on OpenCV
    # clipLimit defines threshold to limit the contrast in case of noise in our image
    # tileGridSize defines the area size in which local equalization will be performed
    clahe_model = cv.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize)

    # For ease of understanding, we explicitly equalize each channel individually
    colorimage_b = clahe_model.apply(colorimage[:,:,0])
    colorimage_g = clahe_model.apply(colorimage[:,:,1])
    colorimage_r = clahe_model.apply(colorimage[:,:,2])

    # Next we stack our equalized channels back into a single image
    colorimage_clahe = np.stack((colorimage_b,colorimage_g,colorimage_r), axis=2)
    
    # returning enhanced image
    return colorimage_clahe
```

```python

def show_images(image, enhanced_image, lable='', figsize=(10,10)):
    """this function will show images"""
    ### defining subplots to plot two images
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=figsize)
    ### ploting original image
    ax1.imshow(image)
    ### setting title for the original image plot
    ax1.set_title(f'Original image')
    ### setting x label to the plot
    ax1.set_xlabel(label)
    
    ### ploting enhanced image
    ax2.imshow(enhanced_image)
    ### setting title to the enhanced image plot
    ax2.set_title(f'CLAHE enhanced image')
    ### setting x label to the plot
    ax2.set_xlabel(label)
    plt.show()
```

```python
### generating 10 random number upto the size of dataset to load random image
random_indices = np.random.randint(0, len(df), 10)
### for each random number
for rand in random_indices:
    ### path of image
    path = df.values[rand][0]
    ### label of image
    label = df.values[rand][1]
    ### loading image
    image = load_image(path)
    ### enhancing image using clahe technique
    enhanced_image = get_enhanced_image(image)
    ### ploting original and enhanced image
    show_images(image, enhanced_image)
```
